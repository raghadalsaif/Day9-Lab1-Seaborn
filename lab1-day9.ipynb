```python
# Import Seaborn
import seaborn as sns
import sqlite3
sns.set()
```

```python
# 9: Draw barplot represents survived people based on sex
sns.barplot(x='sex', y='survived',data=df);
```

```python
# 10: Draw catplot represents survived people based on embarked
g = sns.catplot( data=df, x = "survived", kind="count",col="embarked"  )
```

```python
# 11: Draw boxplot represents distribution of male and female based on age and pclass
plt.figure(figsize=(15,8));
sns.boxplot(x = "sex", y = "age", data=df, hue = "pclass" );
```

```python
# 12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns
group = df[['sibsp', 'parch', 'age', 'fare']]
ax = sns.heatmap(group.corr(), annot = True)
```

```python
# 13: Draw factorplot represents the relation between sibsp and survived columns
sns.factorplot(x ='sibsp', y ='survived', data = df);
```

```python
plt.figure(figsize=(5,8))

sns.violinplot(
    data=df, 
    x="age", 
    y="survived", 
   
    linewidth=2
)

```
